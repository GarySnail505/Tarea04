```python
import scipy.optimize as opt

def f(x):
  return x**3 - 7*x**2 + 14*x - 6

print("a) [0,1] se obtuvo como raíz:",opt.bisect(f, 0, 1, xtol = 10e-2))
print("b) [1,3.2] se obtuvo como raíz:",opt.bisect(f, 1, 3.2, xtol = 10e-2))
print("b) [3,2.4] se obtuvo como raíz:",opt.bisect(f, 3, 2.4, xtol = 10e-2))
```

### Ejercicio 2

```python
## a. Dibuje las gráficas para y = x e y = sin x
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 10, 500)
y1 = x
y2 = np.sin(x)
y3 = 2*np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y1, label='y = x', color='blue')
plt.plot(x, y2, label='y = sin(x)', color='red')
plt.plot(x, y3, label='y = 2sin(x)', color='orange')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.title('Gráficas de y = x e y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-5,5)
plt.show()
```

```python
# b. Use el método de bisección para encontrar soluciones precisas
#   dentro de 10e-5 para el primer valor positivo de x con x = 2 sin x

import scipy.optimize as opt

def f(x):
  return 2 * np.sin(x)

print("Para encontrar el primer raiz postiva es en: ", opt.bisect(f, 2, 4, xtol = 10e-5))
```

### Ejercicio 1

Tenemos en el siguiente ejercicio el siguiente teorema 2.1:
$$|p_n-p| \leq \frac{b-a}{2^n}$$

Reemplazando los valores del ejercicio tenemos:
$$\frac{2-1}{2^n}\leq 10^{-4}$$
$$2^n \geq 10^4$$
$$n \geq 14$$

Estableciendo que:
$$\frac{b-a}{2^n} = \frac{1}{16384}$$
como valor de tolerancia, se buscará la raíz con el siguiente código en Python:

```python
import scipy.optimize as opt

def f(x):
  return x**3 - x - 1

print("La raiz con con el grado de precisión asignada es de:", opt.bisect(f,1,2,xtol=1/16384))
```

```python
import numpy as np
import scipy.optimize as opt

def f(x):
    tol_x = 1e-10
    if abs(x - 1) < tol_x:
        return 0.0
    else:
        return np.sin(np.pi * x)

print("Caso 1 con intervalo: [-0.5 ,2.3] donde a + b < 2:", opt.bisect(f,-0.5 ,2.5 ))
print("Caso 2 con intervalo: [-0.5 ,2.8] donde a + b > 2:", opt.bisect(f,-0.5 ,2.8 ))
print("Caso 2 con intervalo: [-0.11 ,2.11] donde a + b = 2:", opt.bisect(f,-0.11 ,2.11 ))
```
